### **Outlier Analysis, Feature Extraction and Selection/Engineering**

```python
from sklearn.preprocessing import StandardScaler
from scipy import stats
import numpy as np

# 1. Standardization
# We'll standardize the case counts for each county (excluding the Date column)
scaler = StandardScaler()
standardized_data = data.copy()
standardized_data.iloc[:, 1:] = scaler.fit_transform(standardized_data.iloc[:, 1:])

# 2. Outlier Detection
# We'll define outliers as those points which are more than 3 standard deviations from the mean
is_outlier = np.abs(stats.zscore(standardized_data.iloc[:, 1:])) > 3
outlier_counts = np.sum(is_outlier, axis=0)

# Displaying the number of outliers detected for each county
outlier_counts_df = pd.DataFrame(outlier_counts, index=standardized_data.columns[1:], columns=["Outlier Count"])
outlier_counts_df[outlier_counts_df["Outlier Count"] > 0]
```

```python
# Aggregating the data to a monthly frequency
monthly_data = data.set_index('Date').resample('M').sum()

# Recomputing the statistical summary for the monthly data
monthly_statistical_summary = monthly_data.describe()

# Plotting the total cases over time for the same sample counties after aggregation
plt.figure(figsize=(15, 8))
for county in sample_counties:
    plt.plot(monthly_data.index, monthly_data[county], label=county)

plt.title('Monthly Aggregated Chickenpox Cases Over Time in Sample Counties')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.show()

monthly_statistical_summary
```

```python
# Rechecking for outliers in the monthly aggregated data
monthly_standardized_data = monthly_data.copy()
monthly_standardized_data = scaler.fit_transform(monthly_standardized_data)

# Defining outliers as those points which are more than 3 standard deviations from the mean
monthly_is_outlier = np.abs(stats.zscore(monthly_standardized_data)) > 3
monthly_outlier_counts = np.sum(monthly_is_outlier, axis=0)

# Displaying the number of outliers detected for each county in the monthly data
monthly_outlier_counts_df = pd.DataFrame(monthly_outlier_counts, index=monthly_data.columns, columns=["Outlier Count"])
monthly_outlier_counts_df[monthly_outlier_counts_df["Outlier Count"] > 0]
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Decomposing the monthly aggregated time series for Budapest and Baranya
monthly_ts_budapest = monthly_data['BUDAPEST']
monthly_ts_baranya = monthly_data['BARANYA']

decomposition_monthly_budapest = seasonal_decompose(monthly_ts_budapest, model='additive')
decomposition_monthly_baranya = seasonal_decompose(monthly_ts_baranya, model='additive')

# Plotting the decompositions for the monthly data
fig, axes = plt.subplots(ncols=2, nrows=4, sharex=True, figsize=(14, 8))

# Plotting for Budapest
decomposition_monthly_budapest.observed.plot(ax=axes[0,0], title='Budapest (Monthly) - Observed')
decomposition_monthly_budapest.trend.plot(ax=axes[1,0], title='Budapest (Monthly) - Trend')
decomposition_monthly_budapest.seasonal.plot(ax=axes[2,0], title='Budapest (Monthly) - Seasonality')
decomposition_monthly_budapest.resid.plot(ax=axes[3,0], title='Budapest (Monthly) - Residuals')

# Plotting for Baranya
decomposition_monthly_baranya.observed.plot(ax=axes[0,1], title='Baranya (Monthly) - Observed')
decomposition_monthly_baranya.trend.plot(ax=axes[1,1], title='Baranya (Monthly) - Trend')
decomposition_monthly_baranya.seasonal.plot(ax=axes[2,1], title='Baranya (Monthly) - Seasonality')
decomposition_monthly_baranya.resid.plot(ax=axes[3,1], title='Baranya (Monthly) - Residuals')

plt.tight_layout()
plt.show()
```

```python
# Feature Engineering
# Adding lag features - using a 1-month lag as an example
for county in monthly_data.columns:
    monthly_data[county + '_lag1'] = monthly_data[county].shift(1)

# Adding seasonal features (quarter of the year)
monthly_data['quarter'] = monthly_data.index.quarter

# Adding rate of change features (change from the previous month)
for county in monthly_data.columns[:-1]:  # excluding the newly added quarter column
    if 'lag1' not in county:  # excluding the lag columns
        monthly_data[county + '_rate_change'] = monthly_data[county].pct_change()

# Dropping initial rows with NaN values due to lag and rate of change calculations
monthly_data = monthly_data.dropna()

# Displaying the new dataset with engineered features
monthly_data.head()
```

### **Clustering and Anamoly Detection**

```python
# Redefining the range for the number of clusters
range_n_clusters = list(range(2, 11))  # Trying clusters from 2 to 10
original_features = [col for col in monthly_data.columns if 'lag' not in col and 'rate_change' not in col and 'quarter' not in col]
original_monthly_data = monthly_data[original_features]
standardized_original_data = scaler.fit_transform(original_monthly_data)

# Recalculating silhouette scores
silhouette_scores = []
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(standardized_original_data)
    silhouette_avg = silhouette_score(standardized_original_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting silhouette scores to find the optimal number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title("Silhouette Scores for Different Numbers of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
# Performing K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=10)
cluster_labels = kmeans.fit_predict(standardized_original_data)

# Adding the cluster labels back to the original monthly data for interpretation
clustered_data = monthly_data[original_features].copy()
clustered_data['Cluster'] = cluster_labels

# Checking the first few entries of the dataset with the cluster assignments
clustered_data.head()

# Analyzing the clusters
# We can look at the mean case counts for each cluster to understand the characteristics
cluster_means = clustered_data.groupby('Cluster').mean()
cluster_means
```

```python
from sklearn.ensemble import IsolationForest

# Anomaly Detection using Isolation Forest
# We'll detect anomalies in the standardized original data for each county individually

anomalies = {}

for county in original_features:
    # Isolation Forest model
    iso_forest = IsolationForest(contamination=0.01)  # assuming ~1% of the data might be anomalies
    preds = iso_forest.fit_predict(standardized_original_data[:, original_features.index(county)].reshape(-1, 1))

    # Identifying the index of anomalies
    anomaly_indices = np.where(preds == -1)[0]
    anomalies[county] = monthly_data.index[anomaly_indices]

# Displaying anomalies for a few counties as an example
{county: anomalies[county] for county in list(anomalies)[:3]}  # Displaying for first three counties only
```
